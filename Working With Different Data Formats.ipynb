# Reading XML Files

<b>Use the "Card_Txn.xml" file.</b> <br><i>Attributes will be the same as the "Card_Txn.txt" file loaded above.</i>

#### a. Import BeautifulSoup

```python
#use Beautiful Soup
from bs4 import BeautifulSoup
```

#### b. Create Dataframe to hold data

```python
#create Dataframe to hold data
dfcols = ['transaction_id', 'transaction_date', 'cardholder_names', 'card_number', 'merchant', 'amount','merchant_category' ]
df_xml = pd.DataFrame(columns=dfcols)
df_xml
```

#### c. Load xml file and use find_all to get column series for each attribute

```python
#Reads the file
infile = open("../Datasets/Card_Txn.xml","r", encoding = "UTF-8")
contents = infile.read()
```

```python
contents
```

```python
#Parses the XML file
#soup = BeautifulSoup(contents,"xml")
soup = BeautifulSoup(contents,"html.parser")
```

```python
soup
```

```python
# Extracts the data based on tag
trans_ids = soup.find_all('trans_id')
trans_date = soup.find_all('transaction_date')
cardholder_names = soup.find_all('cardholder_name')
card_num = soup.find_all('card_number')
merchant = soup.find_all('merchant')
amount = soup.find_all('amount')
merchant_cat = soup.find_all('merchant_category')
```

```python
# let's find all transaction ids in the list
soup.find_all('trans_id')
```

```python
# print all the cardholder_names column using slice
cardholder_names[0:10]
```

```python
#print first item of record
print(trans_ids[0].get_text())
print(trans_date[0].get_text())
print(cardholder_names[0].get_text())
print(card_num[0].get_text())
print(merchant[0].get_text())
print(amount[0].get_text())
print(merchant_cat[0].get_text())
```

#### d. Use a for loop to iterate through each row, create a temporary row series and append each row to the dataframe prepared in b.

```python
# To append to our dataframe, we need to convert the data to a series with the following code
pd.Series([trans_ids[0].get_text(), trans_date[0].get_text(), cardholder_names[0].get_text(), 
           card_num[0].get_text(), merchant[0].get_text(), amount[0].get_text(), 
           merchant_cat[0].get_text()], index = dfcols)
```

#### Inspect the generated dataframe

```python
# Create a "For Loop" to append all the data to the dataframe we created earlier (df_xml)

for rec in range(0, len(trans_ids)):
    
    row = pd.Series([trans_ids[rec].get_text(), trans_date[rec].get_text(), cardholder_names[rec].get_text(), 
           card_num[rec].get_text(), merchant[rec].get_text(), amount[rec].get_text(), 
           merchant_cat[rec].get_text()], index = dfcols)
    
    df_xml  = df_xml.append(row, ignore_index = True)
```

```python
df_xml.head()
```
